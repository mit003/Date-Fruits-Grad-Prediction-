```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_excel('Date_Fruit_Datasets.xlsx')
# df
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,:-1]
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size=0.3)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_test,y_test)
```

```python
lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, classification_report, accuracy_score
lr_mean = mean_squared_error(lr_pred,y_test)
print(lr_mean)
print(accuracy_score(lr_pred,y_test))
```

# Naive Bayes  

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
nb_pred = nb.predict(X_test)
```

```python
nb_mean = mean_squared_error(nb_pred,y_test)
print(nb_mean)
print(accuracy_score(nb_pred,y_test))
```

# KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
knn_pred = knn.predict(X_test)
```

```python
knn_mean = mean_squared_error(knn_pred,y_test)
print(knn_mean)
print(accuracy_score(knn_pred,y_test))
```

# Suport Vector Machine 

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
```

```python
svm_pred = svm.predict(X_test)
```

```python
svm_mean = mean_squared_error(svm_pred,y_test)
print(svm_mean)
print(accuracy_score(svm_pred,y_test))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
# performing post prunning method
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10)) # size of decision tree chart
plot_tree(dtc,filled=True, fontsize=5)
plt.show()
```

```python
parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','rendom'],
    'max_features':['auto','sqrt','log2'],
    'max_depth':[1,2,3,4,5]
}
```

```python
from sklearn.model_selection import GridSearchCV
dtcv = GridSearchCV(dtc,param_grid=parameter, cv=5, scoring='accuracy')
dtcv.fit(X_train,y_train)
```

```python
dtcv_pred = dtcv.predict(X_test)
```

```python
dtc_pred = dtc.predict(X_test)
```

```python
dtc_mean = mean_squared_error(dtc_pred,y_test)
print("post pumming",dtc_mean)
print("post pumming", accuracy_score(dtc_pred,y_test))
print("\n")
dtcv_mean = mean_squared_error(dtcv_pred, y_test)
print("pre pumming", dtcv_mean)
print("pre pumming", accuracy_score(dtcv_pred,y_test))
```

# Rendom Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=88, criterion='gini')
rfc.fit(X_train,y_train)
```

```python
parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','rendom'],
    'max_features':['auto','sqrt','log2'],
    'max_depth':[1,2,3,4,5]
}
```

```python
from sklearn.tree import plot_tree
 
# Assuming regressor is your trained Random Forest model
# Pick one tree from the forest, e.g., the first tree (index 0)
tree_to_plot = rfc.estimators_[25]
 
# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(tree_to_plot, feature_names=df.columns.tolist(), filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree from Random Forest")
plt.show()
```

```python
rfc_pred = rfc.predict(X_test)
rfc_pred.shape
y.shape
```

```python
rfc_mean = mean_squared_error(rfc_pred,y_test)
print(rfc_mean)
print(accuracy_score(rfc_pred,y_test))
```
